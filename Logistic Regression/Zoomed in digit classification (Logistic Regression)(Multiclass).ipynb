```python
#import necessary libraries
import pandas as pd
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
#load dataset package from scikit learn
digits = load_digits()
```

```python
#preview what is included in this package
dir(digits)
```

```python
#split dataset into 80% training and 20% testing dataset
X_train,X_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
#create the model
model = LogisticRegression()
```

```python
#train the model using training dataset
model.fit(X_train,y_train)
```

```python
#evaluate the model's accuracy using testing dataset
model.score(X_test,y_test)
```

```python
#predict the first 5 images
model.predict(digits.data[0:5])
```

```python
#preview what the model is analyzing (0,1,2,3,4)
for k in range(0,5):
    plt.matshow(digits.images[k])
```

```python
#Get confusion matrix values
y_pred = model.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
cm
```

```python
#visualize confusion matrix
sns.heatmap(cm,annot=True)
plt.xlabel('predicted')
plt.ylabel('truth')
sns.set(rc = {'figure.figsize':(10,8)})
```
